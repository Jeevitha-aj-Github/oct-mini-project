```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the image
img = mpimg.imread('oct retinal b scan.png')

# Convert to grayscale (if its RGB)
if img.ndim == 3:
    img_gray = img.mean(axis=2)
else:
    img_gray = img

# Plot the image again
plt.figure(figsize=(12,5))
plt.imshow(img_gray, cmap='gray')
plt.title("Raw OCT B-scan")
plt.axis('off')
plt.show()

# Extract middle row intensity profile
mid_row = img_gray[img_gray.shape[0]//2, :]

# Plot intensity profile
plt.plot(mid_row, color="black")
plt.title("Intensity Profile Along Central Scan Line")
plt.xlabel("Pixel Position")
plt.ylabel("Intensity")
plt.savefig("oct_central_intensity_profile.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import sobel
import matplotlib.image as mpimg

# 1. Load the OCT image (same file you already used)
img = mpimg.imread("oct retinal b scan.png")

# 2. Convert to grayscale if it is RGB
if img.ndim == 3:
    gray = img.mean(axis=2)
else:
    gray = img

# 3. Apply Sobel edge detection
gx = sobel(gray, axis=1)   # horizontal gradient
gy = sobel(gray, axis=0)   # vertical gradient
edge_mag = np.hypot(gx, gy)

# 4. Normalise for display
edge_mag = edge_mag / edge_mag.max()

# 5. Show original vs edges
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original OCT B-scan")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(edge_mag, cmap='gray')
plt.title("Sobel Edge Map (Retinal Layers)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
